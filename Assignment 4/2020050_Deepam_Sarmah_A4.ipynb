Drop any `NA` values

```python
roo.dropna()
```

```python
X = roo
Y = roo['Suggested Job Role']
X.drop("Suggested Job Role", axis = 1, inplace = True)
```

Implementing a train-test split as in the assignment

```python
trainX, testX, trainY, testY = train_test_split(X, Y, test_size = 0.5, random_state = 0)
```

Implementing an Artificial Neural Network (ANN) using MLP Classifier

```python
standScale = StandardScaler()
trainXScale = standScale.fit_transform(trainX)
testXScale = standScale.transform(testX)
mlp = MLPClassifier(activation = 'relu', hidden_layer_sizes = (300, 200, 100), early_stopping = True, max_iter = 500).fit(trainXScale, trainY)
predictY = mlp.predict(testX)
```

Finding the accuracy

```python
print("The accuracy is ", metrics.accuracy_score(testY, predictY) * 100)
```

Finding the confusion matrix

```python
confusionMat = metrics.confusion_matrix(testY, predictY)
print(confusionMat)
```

Finding the class-wise accuracies

```python
print(confusionMat.diagonal() / confusionMat.sum(axis = 1))
```
